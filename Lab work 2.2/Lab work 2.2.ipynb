Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
sex = input("(f/m): ")
age = input("Bік: ")

if sex not in ['f', 'm'] or not age.isdigit():
    print("Помилка")
else:
    age = int(age)
    if sex == 'f' and age > 18:
        print("Спортсмен-дівчина може брати участь у змаганні")
    elif sex == 'm' and age > 16:
        print("Спортсмен-хлопець може брати участь у змаганні")
    else:
        print("Спортсмен не може брати участь у змаганні")

```

```python
#Виведіть топ-3 слів, що найчастіше зустрічаються#

words = sentence.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

top_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)[:3]

print("Топ-3 слів, які найчастіше зустрічаються:")
for word, count in top_words:
    print(f"{word}: {count} разів")
```
